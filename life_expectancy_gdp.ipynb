```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_data.csv")
```

```python
df
```

```python
# Life expectancy over years
plt.figure(figsize=(12, 8))
sns.lineplot(x="Year", y="Life expectancy at birth (years)", data=df)
plt.xlabel("Year")
plt.ylabel("Life expectancy at birth (years)")
plt.title("Life expectancy over years")
plt.show()
plt.clf()
```

```python
# Has GDP increased over time in the six nations?
plt.figure(figsize=(12, 8))
sns.lineplot(x="Year", y="GDP", hue="Country", marker="o", data=df)
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.title('GDP Over Time in Six Nations')
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x="GDP", y="Life expectancy at birth (years)")
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.title('Scatter Plot of GDP vs. Life Expectancy')
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(12, 8))
mean_life_expectancy = df.groupby("Country")["Life expectancy at birth (years)"].mean().reset_index()
sns.barplot(x="Country", y="Life expectancy at birth (years)", data=mean_life_expectancy)
plt.xticks(rotation=30)
plt.xlabel("Country")
plt.ylabel('Mean Life Expectancy at Birth (years)')
plt.title('Mean Life Expectancy at Birth by Country')
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(12, 8))
plt.title("Comparison of GDP and Life Expectancy Trends by Country Over Time")
ax1 = plt.subplot(1, 2, 1)
sns.lineplot(x="Year", y="GDP", hue="Country", marker="o", data=df)
ax2 = plt.subplot(1, 2, 2)
sns.lineplot(x="Year", y="Life expectancy at birth (years)", hue="Country", marker="o", data=df)
plt.show()
plt.clf()
```
