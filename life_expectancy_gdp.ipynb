```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_data.csv")
```

```python
df
```

```python
# Life expectancy over years
plt.figure(figsize=(12, 8))
sns.lineplot(x="Year", y="Life expectancy at birth (years)", data=df)
plt.xlabel("Year")
plt.ylabel("Life expectancy at birth (years)")
plt.title("Life expectancy over years")
plt.show()
plt.clf()
```

```python
# Has GDP increased over time in the six nations?
plt.figure(figsize=(12, 8))
sns.lineplot(x="Year", y="GDP", hue="Country", marker="o", data=df)
plt.xticks(rotation=90, fontsize=12)
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.title('GDP Over Time in Six Nations')
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x="GDP", y="Life expectancy at birth (years)")
plt.show()
plt.clf()
```
